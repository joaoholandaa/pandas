```python
# Ex01
# Converta a seguinte lista de dados para uma Series Pandas e obtenha:
# Média
# Desvio Padrão
# Máximo Valor

# dados = [10, 20, 42, 9, 12, 35, 24, 10, 8, 14, 21]
```

```python
import pandas as pd
```

```python
dados = [10, 20, 42, 9, 12, 35, 24, 10, 8, 14, 21] 
```

```python
series_dados = pd.Series(dados)
```

```python
media = series_dados.mean()
media
```

```python
desvio = series_dados.std()
desvio
```

```python
maximo = series_dados.max()
maximo
```
